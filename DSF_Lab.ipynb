```python
from scipy import stats
import numpy as np

# Compute Z-scores
z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))

# Set a threshold for identifying outliers
threshold = 3

# Create a boolean mask for filtering out outliers
mask = (z_scores < threshold).all(axis=1)

# Filter out the outliers
df_cleaned = df[mask]

# Display the cleaned DataFrame
print(df_cleaned.head())
```

```python
# Post-cleaning visualization
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_cleaned)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))
threshold = 3
mask = (z_scores < threshold).all(axis=1)
df_cleaned = df[mask]

# Prepare data for feature selection
X = df_cleaned.drop('Smoking Status', axis=1)
y = df_cleaned['Smoking Status']

# Initialize and fit Random Forest model
model = RandomForestClassifier()
model.fit(X, y)

# Get feature importances
importance = model.feature_importances_
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': importance})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

# Display the feature importances
print(feature_importance)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'depression_data.csv'
data = pd.read_csv(file_path)

# Standardize the data (select only numeric columns)
scaler = StandardScaler()
data_standardized = scaler.fit_transform(df.select_dtypes(include=['float64', 'int64']))

# Elbow Method to find optimal k
inertia = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_standardized)  # Use the standardized data
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Define the optimal number of clusters based on the Elbow method
optimal_k = 4  # Replace this with the selected value from the Elbow plot

# Initialize and fit the K-Means model
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(data_standardized)

# `clusters` now contains the cluster label for each data point
df['Cluster'] = clusters  # Optionally add the cluster labels back to the dataset
```

```python
depression_counts = temp_df['Depression_Prediction'].value_counts()
print("Distribution of Depression Prediction Labels:")
print(depression_counts)

import matplotlib.pyplot as plt

depression_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Distribution of Depression Prediction Labels')
plt.xlabel('Depression Prediction')
plt.ylabel('Number of Subjects')
plt.show()
```
